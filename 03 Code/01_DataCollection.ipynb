```python
oecd_df = pd.read_csv("../04 Data/OECD/OECD_integrated.csv")
oecd_df.set_index("TIME_PERIOD", verify_integrity=True, inplace=True, drop=True)
oecd_df.drop(columns=["Unnamed: 0"], inplace=True)
oecd_df.index = pd.to_datetime(oecd_df.index)

oecd_features = pd.read_csv("../04 Data/OECD/OECD_features.csv")
oecd_features.drop(columns=["Unnamed: 0"], inplace=True)

sf_df = pd.read_csv("../04 Data/Salesforce/Salesforce_OpportunityHistory_Timeseries_cleaned.csv")
sf_df.set_index("TIME_PERIOD", verify_integrity=True, inplace=True, drop=True)
sf_df.drop(columns=["index"], inplace=True)
sf_df.index = pd.to_datetime(sf_df.index)

stocks_df = pd.read_csv("../04 Data/Stock Movements Yahoo Finance/Stocks_integrated.csv")
stocks_df.rename(columns={"Date": "TIME_PERIOD"}, inplace=True)
stocks_df.set_index("TIME_PERIOD", verify_integrity=True, inplace=True, drop=True)
stocks_df.index = pd.to_datetime(stocks_df.index)

gt_df = pd.read_csv("../04 Data/Google Trends/GoogleTrends_integrated.csv")
gt_df.set_index("TIME_PERIOD", verify_integrity=True, inplace=True, drop=True)
gt_df.index = pd.to_datetime(gt_df.index)

def filter_columns_by_date(df, cutoff_date):
    # Stelle sicher, dass cutoff_date ein datetime-Objekt ist
    if not isinstance(cutoff_date, pd.Timestamp):
        cutoff_date = pd.to_datetime(cutoff_date)

    # Filtere die Zeilen, die vor dem cutoff_date liegen
    before_cutoff = df[df.index < cutoff_date]

    # Finde die Zeile mit dem aktuellsten Datum vor dem cutoff_date
    if before_cutoff.empty:
        return pd.DataFrame()  # Gib einen leeren DataFrame zur√ºck, wenn keine Zeilen gefunden wurden

    latest_date = before_cutoff.index.max()
    latest_row = before_cutoff.loc[latest_date]

    # Finde Spalten, die in dieser Zeile einen Wert haben
    columns_with_values = [col for col in df.columns if pd.notna(latest_row[col])]

    # Gebe den gefilterten DataFrame zur√ºck, einschlie√ülich der Indexspalte
    return df[columns_with_values]
def find_largest_continuous_timeseries(df, cutoff_date):
    # Filtere den DataFrame basierend auf dem cutoff_date
    filtered_df = filter_columns_by_date(df, cutoff_date)

    # Array zur Speicherung der Anzahl nicht-leerer Werte f√ºr jede Zeile
    non_empty_counts = []

    # Gehe jede Zeile durch und z√§hle die nicht-leeren Werte in den Spalten
    for i in range(len(filtered_df)):
        # Z√§hle die nicht-leeren Werte bis zur aktuellen Zeile mit dem benutzerdefinierten na_filter
        non_empty_count = filtered_df.iloc[:i+1].map(lambda x: (not pd.isna(x)) and (x != "NaN")).sum().sum()
        non_empty_counts.append(non_empty_count)

    # Berechne das Produkt aus Index und Anzahl der nicht-leeren Werte
    products = [index * count for index, count in enumerate(non_empty_counts)]

    # Finde den Index mit dem maximalen Produkt
    max_index = products.index(max(products))

    # Schneide den DataFrame bis zum gefundenen Index ab
    continuous_df = filtered_df.iloc[:max_index+1]

    # Filtere die Spalten, die vom cutoff_date bis zum ermittelten Index Werte haben
    columns_with_values = [
        col for col in continuous_df.columns 
        if continuous_df.loc[cutoff_date:][col].map(lambda x: (not pd.isna(x)) and (x != "NaN")).all()
    ]

    # Gebe den gefilterten DataFrame einschlie√ülich der Indexspalte zur√ºck
    return continuous_df[columns_with_values]
def form_largest_clean_dataset(sources_df, cutoff_end_date='2024-12-31', cutoff_start_date="2018-01", index_col="TIME_PERIOD"):
    cutoff_end_date = pd.to_datetime(cutoff_end_date)
    sources_df = sources_df.apply(preprocess_series)
    sources_df.sort_index(ascending=False, inplace=True)
    sources_df = sources_df[(sources_df.index >= pd.to_datetime(cutoff_start_date)) & (cutoff_end_date >= sources_df.index)]# & (sources_df['TIME_PERIOD'] < "2024-09")]
    sources_df = find_largest_continuous_timeseries(sources_df, cutoff_end_date)

    sources_df.info()
    sources_df.index.name = index_col
    sources_df.reset_index(inplace=True)
    display(sources_df.head())
    print(sources_df.shape)
    print(f"= {sources_df.shape[0]*sources_df.shape[1]}")
    plot_missing_values(sources_df)
    return sources_df
def print_columns(columns, oecd_features, col_tag="Removed"):
    description = feature_descriptions(oecd_features[oecd_features["Feature"].isin(columns)])
    print(f"{col_tag} Columns: In total", len(oecd_features[oecd_features["Feature"].isin(columns)]), "Features from OECD")
    for key, info in description.items():
        #print(info["Name"])
        print(info["Dataset"][1], info["Description"])
    print(f"Other {col_tag} Features:")
    for c in columns:
        if c not in oecd_features["Feature"].values:
            print(c)
def compare_columns(original_df, cleaned_df, oecd_features):
    original_columns = set(original_df.columns)
    cleaned_columns = set(cleaned_df.columns)
    
    removed_columns = original_columns - cleaned_columns
    retained_columns = cleaned_columns.intersection(original_columns)

    print_columns(removed_columns, oecd_features, col_tag="Removed")
    print_columns(retained_columns, oecd_features, col_tag="Retained")

    return removed_columns, retained_columns
def process_datasets(oecd_df, sf_df, stocks_df, gt_df, oecd_features, cutoff_end_date, cutoff_start_date, comparison_index="end"):
    oecd_df_cleaned = form_largest_clean_dataset(oecd_df, cutoff_end_date=cutoff_end_date, cutoff_start_date=cutoff_start_date)
    compare_columns(oecd_df, oecd_df_cleaned, oecd_features)

    oecd_sf_df_cleaned = form_largest_clean_dataset(pd.concat([oecd_df, sf_df], axis=1), cutoff_end_date=cutoff_end_date, cutoff_start_date=cutoff_start_date)
    compare_columns(pd.concat([oecd_df, sf_df], axis=1), oecd_sf_df_cleaned, oecd_features)

    oecd_sf_stocks_df_cleaned = form_largest_clean_dataset(pd.concat([oecd_df, sf_df, stocks_df], axis=1), cutoff_end_date=cutoff_end_date, cutoff_start_date=cutoff_start_date)
    compare_columns(pd.concat([oecd_df, sf_df, stocks_df], axis=1), oecd_sf_stocks_df_cleaned, oecd_features)

    oecd_sf_stocks_gt_df_cleaned = form_largest_clean_dataset(pd.concat([oecd_df, sf_df, stocks_df, gt_df], axis=1), cutoff_end_date=cutoff_end_date, cutoff_start_date=cutoff_start_date)
    compare_columns(pd.concat([oecd_df, sf_df, stocks_df, gt_df], axis=1), oecd_sf_stocks_gt_df_cleaned, oecd_features)

    comparison_index = [cutoff_end_date] if comparison_index == "end" else [cutoff_start_date]
    # Create a comparison DataFrame for this cutoff date
    comparison = pd.DataFrame({
        # "OECD": [oecd_df_cleaned.shape[0] * oecd_df_cleaned.shape[1]],
        # "OECD_SF": [oecd_sf_df_cleaned.shape[0] * oecd_sf_df_cleaned.shape[1]],
        # "OECD_SF_SM": [oecd_sf_stocks_df_cleaned.shape[0] * oecd_sf_stocks_df_cleaned.shape[1]],
        # "OECD_SF_SM_GT": [oecd_sf_stocks_gt_df_cleaned.shape[0] * oecd_sf_stocks_gt_df_cleaned.shape[1]]
        "OECD": [oecd_df_cleaned.shape[1]],
        "OECD_SF": [oecd_sf_df_cleaned.shape[1]],
        "OECD_SF_SM": [oecd_sf_stocks_df_cleaned.shape[1]],
        "OECD_SF_SM_GT": [oecd_sf_stocks_gt_df_cleaned.shape[1]]
    }, index=comparison_index)

    return oecd_df_cleaned, oecd_sf_df_cleaned, oecd_sf_stocks_df_cleaned, oecd_sf_stocks_gt_df_cleaned, comparison
# Define range of cutoff dates
cutoff_dates = pd.date_range(start="2024-06-30", end="2024-12-31", freq='M')

def dataset_size_experiment(oecd_df, sf_df, stocks_df, gt_df, oecd_features, cutoff_dates=pd.date_range(start="2024-06", end="2024-12", freq='MS'), start_or_end="end", other_end_date_fixed="2018-01"):
    # Initialize an empty DataFrame to aggregate comparisons
    aggregate_comparison = pd.DataFrame()

    oecd_retained_cols = None
    oecd_removed_cols = {}
    last_oecd_df = None
    cutoff_dates = cutoff_dates if start_or_end == "end" else reversed(cutoff_dates)
    # Process datasets for each cutoff date and aggregate results
    first_iteration = True
    for cutoff_date in cutoff_dates:
        if start_or_end == "end":
            oecd_df_cleaned_i, _, _, _, comparison = process_datasets(oecd_df, sf_df, stocks_df, gt_df, oecd_features, cutoff_end_date=cutoff_date, cutoff_start_date=other_end_date_fixed, comparison_index=start_or_end)
        else:
            oecd_df_cleaned_i, _, _, _, comparison = process_datasets(oecd_df, sf_df, stocks_df, gt_df, oecd_features, cutoff_start_date=cutoff_date, cutoff_end_date=other_end_date_fixed, comparison_index=start_or_end)
        if first_iteration == True:
            oecd_removed_cols[cutoff_date], oecd_retained_cols = compare_columns(oecd_df, oecd_df_cleaned_i, oecd_features)
        else:
            oecd_removed_cols[cutoff_date], _ = compare_columns(oecd_df_cleaned_i_last, oecd_df_cleaned_i, oecd_features)
        oecd_df_cleaned_i_last = oecd_df_cleaned_i
        aggregate_comparison = pd.concat([aggregate_comparison, comparison])
        first_iteration = False

    return aggregate_comparison, oecd_retained_cols, oecd_removed_cols
def plot_comparison(comparison_df, oecd_retained_cols, oecd_removed_cols, oecd_features):
    comparison_df.plot(kind='bar', figsize=(12, 7))
    plt.title('Dataset Size Comparison Across Different Cutoff Dates')
    plt.ylabel('Dataset Size (Number of Elements)')
    plt.xlabel('Cutoff Date')
    plt.xticks(rotation=45)
    plt.legend(title='Dataset Type')
    plt.tight_layout()
    plt.show()

    print("Retained OECD Columns:")
    print_columns(oecd_retained_cols, oecd_features, col_tag="Retained")
    print("Removed OECD Columns (cumulated over cutoff date):")
    for i, cols in oecd_removed_cols.items():
        print_columns(cols, oecd_features, col_tag=f"Removed {i}")
```

```python
def convert_string_to_float(value):
    try:
        # Ersetze Komma durch Punkt und konvertiere in Float
        return float(value.replace('.','').replace(',', '.').replace(' ', ''))
    except (ValueError, AttributeError):
        return 0.0  # Gebe 0.0 zur√ºck, wenn die Konvertierung fehlschl√§gt
def generate_monthly_stage_data(df):
    # Konvertiere Datumsfelder in datetime-Objekte
    df['Opportunity Close Date'] = pd.to_datetime(df['Opportunity Close Date'], format='%d.%m.%Y', errors='coerce', dayfirst=True)
    df['Last Stage Change Date'] = pd.to_datetime(df['Last Stage Change Date'], format='%d.%m.%Y', errors='coerce', dayfirst=True)
    df['Last Modified'] = pd.to_datetime(df['Last Modified'], format='%d.%m.%Y', errors='coerce', dayfirst=True)
    df['Created Date'] = pd.to_datetime(df['Created Date'], format='%d.%m.%Y', errors='coerce', dayfirst=True)
    df['Close Date'] = pd.to_datetime(df['Close Date'], format='%d.%m.%Y', errors='coerce', dayfirst=True)

    # Die geschlossenen Stages
    closed_stages = ['Closed Won', 'Closed Cancelled', 'Closed Lost']

    # Konvertiere gesch√§tzte Werte in Float
    #df['Estimated Value in OC (converted)'] = df['Estimated Value in OC (converted)'].apply(convert_string_to_float)
    
    # Liste f√ºr die neuen Daten
    records = []

    # Iteriere √ºber jede Opportunity
    for opportunity_name, group in df.groupby('Opportunity ID'):
        # Sortiere die Gruppe nach dem Datum der letzten Stage-√Ñnderung
        group = group.sort_values(by='Last Modified')
        
        # Erhalte das Erstellungsdatum der Opportunity
        created_date = group['Created Date'].min()
        
        # Bestimme die letzte Stage in der Opportunity
        last_stage = group['To Stage'].iloc[-1]

        # Verwende das Erstellungsdatum der Opportunity als Startdatum
        if pd.isnull(created_date):
            continue
        start_date = created_date
        
        # Verwende das Opportunity Close Date, falls vorhanden, sonst das aktuelle Datum als Enddatum
        end_date = (group['Opportunity Close Date'].max() if pd.notnull(group['Opportunity Close Date']).any() and group['Opportunity Close Date'].max() <= datetime.now() else datetime.now())

        # Gehe durch jeden Monat von start_date bis end_date
        current_date = start_date
        previous_stage = None
        previous_commitment = False
        previous_closed = False
        previous_won = False

        while current_date <= end_date:
            # W√§hle den passenden Eintrag aus der Gruppe
            current_row = group.loc[group['Last Modified'] <= current_date].sort_values(by='Last Modified').tail(1)
            if current_row.empty:
                current_row = group.loc[group['Last Modified'] >= current_date].sort_values(by='Last Modified').head(1)
            
            if current_row.empty:
                current_date = (current_date + pd.offsets.MonthEnd(1)) + pd.offsets.MonthBegin(1)
                continue
            
            row = current_row.iloc[0]

            # Berechnung der zus√§tzlichen Felder
            is_new = current_date.strftime('%Y-%m') == created_date.strftime('%Y-%m')
            is_closed_this_month = not previous_closed and row['Closed'] == 1
            is_won_this_month = not previous_won and row['Won'] == 1
            new_release_for_commitment = not previous_commitment and row['Release For Commitment'] == 1
            new_stage = row['To Stage'] if previous_stage and previous_stage != row['To Stage'] else None
            expected_amount = float(row['Estimated Value in OC (converted)']) * int(row['Probability (%)']) / 100 if pd.notnull(row['Estimated Value in OC (converted)']) else 0

            records.append({
                'Opportunity ID': row['Opportunity ID'],
                'Opportunity Name': row['Opportunity Name'],
                'Year-Month': current_date.strftime('%Y-%m'),
                'Stage': row['To Stage'],
                'Stage Duration': row['Stage Duration'],
                'Estimated Value': row['Estimated Value in OC (converted)'],
                'Probability': row['Probability (%)'],
                'Expected Amount': expected_amount,
                'Close Date': row['Close Date'],
                'Decision Date': row['Decision Date'],
                'Release for Commitment': row['Release For Commitment'],
                'New': is_new,
                'New Release for Commitment this Month': new_release_for_commitment,
                'New Stage': new_stage,
                'Closed This Month': is_closed_this_month,
                'Won This Month': is_won_this_month
            })

            # Update previous values for next iteration
            previous_stage = row['To Stage']
            previous_commitment = row['Release For Commitment'] == 1
            previous_closed = row['Closed'] == 1
            previous_won = row['Won'] == 1

            # Wenn die Stage eine geschlossene Stage und die letzte Stage ist, brich die Schleife ab
            if row['To Stage'] in closed_stages and row['To Stage'] == last_stage:
                break

            # Gehe zum n√§chsten Monat
            current_date = (current_date + pd.offsets.MonthEnd(1)) + pd.offsets.MonthBegin(1)

    # Erstelle einen neuen DataFrame aus den gesammelten Datens√§tzen
    monthly_stage_df = pd.DataFrame(records)
    return monthly_stage_df

monthly_opportunities_stage_df = generate_monthly_stage_data(opportunity_history_report_df)
monthly_opportunities_stage_df.info()
monthly_opportunities_stage_df.head()
monthly_opportunities_stage_df.to_excel("../04 Data/Salesforce/Monthly_opportunities_stage_df.xlsx")
```

```python
def analyze_date_range(date_list):
    # Konvertiere die Liste von Strings in datetime-Objekte
    dates = pd.to_datetime(date_list, format='%Y-%m', errors='coerce')
    
    # Filtere ung√ºltige Datumswerte (NaT)
    valid_dates = dates.dropna()

    # Berechne die L√§nge der Liste
    list_length = len(valid_dates)
    
    if list_length == 0:
        return {
            'length': 0,
            'min_date': None,
            'max_date': None,
            'month_difference': 0
        }

    # Bestimme den kleinsten und gr√∂√üten Wert
    min_date = valid_dates.min()
    max_date = valid_dates.max()

    # Berechne die Anzahl der Monate zwischen dem kleinsten und gr√∂√üten Wert
    month_difference = (max_date.year - min_date.year) * 12 + (max_date.month - min_date.month)

    return {
        'length': list_length,
        'min_date': min_date.strftime('%Y-%m'),
        'max_date': max_date.strftime('%Y-%m'),
        'month_difference': month_difference
    }

def generate_summary_report(df):
    #current_month = "2024-10"#datetime.now().strftime('%Y-%m')

    # Helper function to compute metrics
    def compute_metrics(group):
        count = group.shape[0]
        total_amount = group['Estimated Value'].sum()
        avg_probability = group['Probability'].mean()
        total_expected_amount = group['Expected Amount'].sum()
        return count, total_amount, avg_probability, total_expected_amount

    def metrics_to_columns(report, metrics_group_name, metrics_list):
        if len(metrics_list) < 1:
            metrics_list = [0, 0.0, 0.0, 0.0]
        report[f'{metrics_group_name} Count'] = metrics_list[0]
        report[f'{metrics_group_name} Amount'] = metrics_list[1]
        report[f'{metrics_group_name} Avg Probability'] = metrics_list[2]
        report[f'{metrics_group_name} Expected Amount'] = metrics_list[3]
        return report

    def metrics_to_columns_dict(report, metrics_group_name, metrics_dict):
        for stage, metrics in metrics_dict.items():
            report = metrics_to_columns(report, f"{metrics_group_name} {stage}", metrics)
        return report

    # Metrics for next 12 months
    def future_metrics(fm_df, date_col):
        metrics = {}
        fm_df = fm_df[fm_df[date_col].notnull()].copy()
        for i in range(1, 13):
            future_date = datetime.now() + pd.DateOffset(months=i)
            future_year = future_date.year
            future_month = future_date.month
            future_df = fm_df[(fm_df[date_col].dt.year == future_year) & (fm_df[date_col].dt.month == future_month)]
            metrics[f"in {i}"] = compute_metrics(future_df)
            #metrics_to_columns(f'{date_col} in {i}', compute_metrics(future_df))
        return metrics
    
    results_df = None

    for current_month in df["Year-Month"].unique().tolist():

        # Total and per stage metrics (except closed)
        current_df = df[df['Year-Month'] == current_month]
        current_df_stage_filtered = current_df.dropna(subset=['Stage'])
        current_df_newstage_filtered = current_df.dropna(subset=['New Stage'])
        open = current_df[~current_df['Stage'].str.startswith('Closed')] # Offene Opportunities sind die, die diesen Monat einen Eintrag haben deren Stage keine Closed Stage ist.
        new = current_df[current_df['New'] == 1]
        closed_this_month = current_df[current_df['Closed This Month']]
        won_this_month = current_df[current_df['Won This Month']]
        commitment = current_df[(current_df['Release for Commitment'] == 1) & (~current_df['Stage'].str.startswith('Closed'))]

        avg_stage_duration = open['Stage Duration'].mean()
        avg_amount_open = open['Estimated Value'].mean()
        avg_probability_open = open['Probability'].mean()

        open_metrics = compute_metrics(open)
        new_metrics = compute_metrics(new)
        won_metrics = compute_metrics(won_this_month)
        closed_metrics = compute_metrics(closed_this_month)
        commitment_metrics = compute_metrics(commitment)

        stage_metrics = None
        if not current_df_stage_filtered.empty:
            stage_metrics = current_df_stage_filtered.groupby('Stage').apply(compute_metrics, include_groups=False).to_dict()
        new_stage_metrics = None
        if not current_df_newstage_filtered.empty:
            new_stage_metrics = current_df_newstage_filtered.groupby('New Stage').apply(compute_metrics, include_groups=False).to_dict()

        commitment_date_metrics = future_metrics(commitment, 'Decision Date')
        close_date_metrics = future_metrics(open, 'Close Date')
        decision_date_metrics = future_metrics(open, 'Decision Date')

        # Compile the report
        report = {
            'TIME_PERIOD': current_month,
            'Avg Stage Duration Open': avg_stage_duration,
            'Avg Amount Open': avg_amount_open
            #'Avg Probability Open': avg_probability_open
        }

        metrics_to_columns(report, "Open", open_metrics)
        metrics_to_columns(report, "New", new_metrics)
        metrics_to_columns(report, "Closed", closed_metrics)
        metrics_to_columns(report, "Won", won_metrics)
        metrics_to_columns(report, "Commitment", commitment_metrics)
            
        # Add stage-specific metrics to the report
        if stage_metrics:
            metrics_to_columns_dict(report, "Stage", stage_metrics)
        # Add new opportunities metrics per stage to the report
        if new_stage_metrics:
            metrics_to_columns_dict(report, "New Stage", new_stage_metrics)

        # Add future metrics to the report
        metrics_to_columns_dict(report, "Commitment", commitment_date_metrics)
        metrics_to_columns_dict(report, "Close Date", close_date_metrics)
        metrics_to_columns_dict(report, "Decision Date", decision_date_metrics)

        if results_df is None:
            results_df = pd.DataFrame([report])
        else:
            results_df = pd.concat([results_df, pd.DataFrame([report])])

    # Convert TIME_PERIOD to datetime and sort in descending order
    results_df['TIME_PERIOD'] = pd.to_datetime(results_df['TIME_PERIOD'], format='%Y-%m')
    results_df.sort_values(by='TIME_PERIOD', ascending=False, inplace=True)

    return results_df

monthly_opportunities_stage_df = pd.read_excel("../04 Data/Salesforce/Monthly_opportunities_stage_df.xlsx")
monthly_opportunities_stage_df['Close Date'] = pd.to_datetime(monthly_opportunities_stage_df['Close Date'], format='%d.%m.%Y', errors='coerce', dayfirst=True)
monthly_opportunities_stage_df['Decision Date'] = pd.to_datetime(monthly_opportunities_stage_df['Decision Date'], format='%d.%m.%Y', errors='coerce', dayfirst=True)

sf_report_df = generate_summary_report(monthly_opportunities_stage_df)
sf_report_df.info()
display(sf_report_df.head())

months = sf_report_df["TIME_PERIOD"].unique().tolist()
result = analyze_date_range(months)
print(result)
sf_report_df.columns.tolist()

sf_report_df.to_excel("../04 Data/Salesforce/Salesforce_OpportunityHistory_Timeseries.xlsx", index=False)
sf_report_df.to_csv("../04 Data/Salesforce/Salesforce_OpportunityHistory_Timeseries.csv", index=False)

# Funktion zum Ersetzen von Ausrei√üern
def replace_outliers(series, factor):
    return series.where(
        ~((series > series.shift(1) * factor) & (series > series.shift(-1) * factor)),
        (series.shift(1) + series.shift(-1)) / 2
    )

sf_report_df.set_index("TIME_PERIOD", inplace=True)
sf_report_filtert_and_outliers_removed_df = sf_report_df.apply(replace_outliers, factor=1e3)[["New Amount", "Open Expected Amount", "Closed Amount", "Won Amount", "Open Amount", "Won Count"]]
sf_report_filtert_and_outliers_removed_df.reset_index(inplace=True)

sf_report_filtert_and_outliers_removed_df.to_excel("../04 Data/Salesforce/Salesforce_OpportunityHistory_Timeseries_cleaned.xlsx", index=False)
sf_report_filtert_and_outliers_removed_df.to_csv("../04 Data/Salesforce/Salesforce_OpportunityHistory_Timeseries_cleaned.csv", index=False)
```

```python
def merge_csv_files(input_directory, index_col="Monat"):
    dataframes = []
    
    for filename in os.listdir(input_directory):
        if filename.endswith('.csv'):
            file_path = os.path.join(input_directory, filename)
            df = pd.read_csv(file_path, index_col=index_col)
            dataframes.append(df)
    
    merged_df = pd.concat(dataframes, axis=1)
    merged_df.index = pd.to_datetime(merged_df.index, format='%Y-%m')
    return merged_df

def integrate_googletrends(input_directory = "../04 Data/Google Trends/Raw", output_file = "../04 Data/Google Trends/GoogleTrends_integrated.csv", column_prefix = "GT ", index_raw="Monat", index_output="TIME_PERIOD"):
    gt_df = merge_csv_files(input_directory, index_col=index_raw)
    gt_df.rename(columns={col: column_prefix + col.replace(": (Weltweit)", "") for col in gt_df.columns}, inplace=True)
    gt_df.index.name = index_output
    gt_df.sort_index(ascending=False, inplace=True)
    gt_df.replace("<1", "0", inplace=True)
    gt_df = gt_df.apply(pd.to_numeric, errors='coerce').fillna(0).astype(int)
    gt_df.to_csv(output_file)
    return gt_df

gt_df = integrate_googletrends()
gt_df.info()
gt_df.head()
```
